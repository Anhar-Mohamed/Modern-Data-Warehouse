```python
import kagglehub
import os
import pandas as pd

# Download latest version
path = kagglehub.dataset_download("mohamedmagdy11/egypt-gold-prices-daily-updated")

print("Path to dataset files:", path)
# List all files in the dataset directory
print("Files in dataset directory:", os.listdir(path))
```

```python
data = pd.read_csv(os.path.join(path, "/kaggle/input/egypt-gold-prices-daily-updated/data.csv"))
data.head()
```

```python
# Drop columns from index 19 to the end]

cols_to_drop = data.columns[19:]
print(cols_to_drop)

```

```python
data = data.drop(columns=cols_to_drop)
```

#unpivoting my dataset

```python
# Columns to keep as identifiers
id_vars = ['Date']

# Columns to unpivot
value_vars = [col for col in data.columns if col != 'Date']

# Melt the dataframe
melted = data.melt(id_vars=id_vars, value_vars=value_vars,
                   var_name='Karat_Type', value_name='Price')

# Split 'Karat_Type' into Karat and Price_Type
melted[['Karat', 'Price_Type']] = melted['Karat_Type'].str.extract(r'(\d+K) - (.*)')
```

```python
# reorder columns
melted = melted[['Date', 'Karat', 'Price_Type', 'Price']]

# Sort by Date and Karat
melted = melted.sort_values(['Date', 'Karat']).reset_index(drop=True)

melted.head()
```

```python
melted['Date']=pd.to_datetime(melted['Date'])
extra_inf['Date']=pd.to_datetime(extra_inf['Date'])
```

#marge two tables to make full data overview

```python
full_data=pd.merge(melted,extra_inf,on='Date',how='left')
```

```python
full_data.head()
```

```python
full_data.info()
```

# Split dataset based on Price_Type

```python
global_data = full_data[full_data['Price_Type'] == 'Global Price'].copy()
local_data = full_data[full_data['Price_Type'] != 'Global Price'].copy()

# Optional: rename columns for clarity
global_data = global_data.rename(columns={'Price': 'Global_Price', 'EGP=X': 'EGP_USD'})
local_data = local_data.rename(columns={'Price': 'Local_Price', 'EGP=X': 'EGP_USD'})

# Reset index for both DataFrames
global_data = global_data.reset_index(drop=True)
local_data = local_data.reset_index(drop=True)

# Check shapes
print("Global data shape:", global_data.shape)
print("Local data shape:", local_data.shape)

```

# Extract type of Price_Type

```python
# Extract Buy/Sell from Price_Type directly
local_data['Type'] = local_data['Price_Type'].str.extract(r'Local Price/(Buy|Sell)')

# Drop the old Price_Type column
local_data.drop(columns=['Price_Type'], inplace=True)

# Reorder columns (optional)
local_data = local_data[['Date', 'Karat', 'Type', 'Local_Price', 'EGP_USD']]

# Check result
local_data.head()
```
